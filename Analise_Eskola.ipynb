```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a funo com a singularidade
def f(x):
    return 1 / x

# Definindo o intervalo de x
x_values = np.linspace(-2, 2, 400)  # Pontos de -2 a 2

# Calculando os valores de f(x)
y_values = f(x_values)

# Plotando a funo
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=r'$f(x) = \frac{1}{x}$', color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Funo com Singularidade em $x = 0$')
plt.axvline(x=0, color='red', linestyle='--', label='Singularidade em $x=0$')
plt.legend()
plt.grid(True)
plt.show()
```
